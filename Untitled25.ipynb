```python
# 라이브러리 임포트 및 설정
import os
import json
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization, Dropout, Input, Concatenate
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from google.colab import drive
from tensorflow.keras.applications import EfficientNetB0  # EfficientNetB0로 변경
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import keras_tuner as kt
import unicodedata
import matplotlib.font_manager as fm
import matplotlib
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.utils import Sequence
```

```python
# Google Drive 마운트
drive.mount('/content/drive')

# 데이터 경로 설정
base_dir = '/content/drive/MyDrive/Colab Notebooks/testset'
image_dir = os.path.join(base_dir, 'TS_Color_위해물품')
label_dir = os.path.join(base_dir, 'TL_위해물품')

# 밀도 정보 및 색상 정보 정의
density_info = {
    "metal_density_high": 8.0,    # 금속의 높은 밀도 기준 값 (예시)
    "metal_density_low": 5.0,     # 금속의 낮은 밀도 기준 값 (예시)
    "plastic_density_high": 1.5,  # 플라스틱의 높은 밀도 기준 값 (예시)
    "plastic_density_low": 0.8,   # 플라스틱의 낮은 밀도 기준 값 (예시)
    "metal_high_density_color": "#00008B",    # 짙은 파란색 (DarkBlue)
    "metal_low_density_color": "#90EE90",     # 연두색 (LightGreen)
    "plastic_high_density_color": "#FF8C00",  # 짙은 주황색 (DarkOrange)
    "plastic_low_density_color": "#FFA07A",   # 옅은 주황색 (LightSalmon)
}

# JSON 파일의 메타데이터 업데이트
directory_path = label_dir

# 디렉토리 내 모든 JSON 파일을 순회
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)

        # JSON 파일 읽기
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

        # 메타데이터 업데이트
        if "meta" in data:
            for item in data["meta"]:
                if "material_inside" in item:
                    # 밀도값 추가 (실제 값으로 설정)
                    if item["material_inside"] == "Metal":
                        # 아이템의 실제 금속 밀도 값을 설정해야 합니다.
                        # 예시로 임의의 값을 사용합니다.
                        item["density"] = 7.5  # 실제 밀도 값으로 대체
                        # 밀도에 따른 색상 결정
                        if item["density"] >= density_info["metal_density_high"]:
                            item.update({
                                "color": density_info["metal_high_density_color"]
                            })
                        else:
                            item.update({
                                "color": density_info["metal_low_density_color"]
                            })
                    elif item["material_inside"] == "Plastic":
                        # 아이템의 실제 플라스틱 밀도 값을 설정해야 합니다.
                        # 예시로 임의의 값을 사용합니다.
                        item["density"] = 1.2  # 실제 밀도 값으로 대체
                        # 밀도에 따른 색상 결정
                        if item["density"] >= density_info["plastic_density_high"]:
                            item.update({
                                "color": density_info["plastic_high_density_color"]
                            })
                        else:
                            item.update({
                                "color": density_info["plastic_low_density_color"]
                            })

        # 변경된 데이터 저장
        with open(file_path, 'w', encoding='utf-8') as file:
            json.dump(data, file, ensure_ascii=False, indent=4)

print("메타데이터가 모든 파일에 업데이트되었습니다.")
```

```python
# 이미지에서 밀도값 추정 함수 정의
def estimate_density_from_image(image):
    # 이미지의 평균 밝기 계산 (예시)
    grayscale_image = tf.image.rgb_to_grayscale(image)
    mean_intensity = tf.reduce_mean(grayscale_image)
    # 밀도값은 밝기에 반비례한다고 가정 (예시)
    # 밝기가 높을수록 밀도가 낮은 물질로 간주
    estimated_density = 1.0 / (mean_intensity + 1e-5)  # 0으로 나누기 방지
    return estimated_density.numpy()

# 이미지 경로와 추정된 밀도값을 저장할 리스트 초기화
image_paths = []
labels = []
estimated_densities = []

# 클래스 레이블 가져오기 및 Unicode 정규화
class_labels = sorted(os.listdir(image_dir))
class_labels = [unicodedata.normalize('NFC', label) for label in class_labels]
num_classes = len(class_labels)
print(f"클래스 레이블: {class_labels}")

# 이미지와 레이블 수집 및 밀도값 추정
for class_label in class_labels:
    class_dir = os.path.join(image_dir, class_label)
    if os.path.isdir(class_dir):
        for img_name in os.listdir(class_dir):
            if img_name.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
                img_path = os.path.join(class_dir, img_name)
                image_paths.append(img_path)
                labels.append(unicodedata.normalize('NFC', class_label))
                # 이미지 로드 및 전처리
                image = tf.io.read_file(img_path)
                image = tf.image.decode_image(image, channels=3)
                image = tf.image.resize(image, [224, 224])
                image = image / 255.0  # 정규화
                # 밀도값 추정
                density = estimate_density_from_image(image)
                estimated_densities.append(density)

# 데이터를 데이터프레임으로 구성
data_df = pd.DataFrame({'filename': image_paths, 'label': labels, 'density': estimated_densities})

# 밀도값 정규화 (모델 입력에 사용하기 위해)
scaler = MinMaxScaler()
data_df['density_normalized'] = scaler.fit_transform(data_df[['density']])

# 레이블 인코딩 (클래스 가중치 계산을 위해 정수 인코딩)
label_encoder = LabelEncoder()
data_df['label_encoded'] = label_encoder.fit_transform(data_df['label'])
```

```python
# 데이터세트 분할 (70% 훈련, 20% 검증, 10% 테스트)
train_df, temp_df = train_test_split(data_df, test_size=0.3, random_state=42, stratify=data_df['label'])
val_df, test_df = train_test_split(temp_df, test_size=0.33, random_state=42, stratify=temp_df['label'])  # 0.33 * 0.3 ≈ 0.1

# 인덱스 리셋
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# 클래스 가중치 계산
classes = np.unique(train_df['label_encoded'])
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=train_df['label_encoded']
)

# 최대 가중치 제한 적용
max_weight = 5.0  # 최대 가중치 제한
class_weights = np.clip(class_weights, 0, max_weight)

# 클래스 가중치 딕셔너리 생성
class_weight_dict = dict(zip(classes, class_weights))

print("클래스 가중치:")
print(class_weight_dict)
```

```python
# 배치 크기 설정
batch_size = 32

# 데이터 제너레이터 설정
datagen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest'
)
datagen_val_test = ImageDataGenerator(rescale=1./255)

# 커스텀 제너레이터 클래스 정의
class CustomDataGenerator(Sequence):
    def __init__(self, df, batch_size, is_training, datagen):
        self.df = df
        self.batch_size = batch_size
        self.is_training = is_training
        self.datagen = datagen
        self.on_epoch_end()
    def __len__(self):
        return int(np.ceil(len(self.df) / self.batch_size))
    def __getitem__(self, idx):
        batch_df = self.df.iloc[idx * self.batch_size : (idx + 1) * self.batch_size]
        batch_images = []
        for filename in batch_df['filename']:
            img = tf.keras.preprocessing.image.load_img(filename, target_size=(224, 224))
            img = tf.keras.preprocessing.image.img_to_array(img)
            batch_images.append(img)
        batch_images = np.array(batch_images)
        if self.is_training:
            batch_images = self.datagen.flow(batch_images, batch_size=len(batch_images), shuffle=False).next()
        else:
            batch_images = batch_images / 255.0
        batch_density = batch_df['density_normalized'].values
        batch_density = batch_density.reshape(-1, 1)  # 추가됨
        batch_labels = batch_df['label']
        batch_labels_encoded = label_encoder.transform(batch_labels)
        batch_labels_onehot = tf.keras.utils.to_categorical(batch_labels_encoded, num_classes=num_classes)
        return [batch_images, batch_density], batch_labels_onehot
    def on_epoch_end(self):
        if self.is_training:
            self.df = self.df.sample(frac=1).reset_index(drop=True)
```

```python
# 제너레이터 생성
train_generator = CustomDataGenerator(train_df, batch_size, is_training=True, datagen=datagen_train)
validation_generator = CustomDataGenerator(val_df, batch_size, is_training=False, datagen=datagen_val_test)
test_generator = CustomDataGenerator(test_df, batch_size, is_training=False, datagen=datagen_val_test)

steps_per_epoch = len(train_generator)
validation_steps = len(validation_generator)
test_steps = len(test_generator)
```

```python
# 클래스 인덱스와 클래스명 매핑
class_indices = {v: k for k, v in enumerate(label_encoder.classes_)}
inv_class_indices = {v: k for k, v in class_indices.items()}

# 클래스 레이블 매핑 생성 (한글 클래스명 간소화)
class_label_mapping = {
    'TS_Color_위해물품_Bullet_총알': '총알',
    'TS_Color_위해물품_Card knife_카드칼(접은 형태)(167)': '카드칼',
    'TS_Color_위해물품_Electroshock weapon_전자충격기': '전자충격기',
    'TS_Color_위해물품_Grenade_수류탄': '수류탄',
    'TS_Color_위해물품_Knife blade_칼날(030)': '칼날',
    'TS_Color_위해물품_Knife-D_맥가이버 칼(접은 형태)(029)': '맥가이버 칼',
    'TS_Color_위해물품_Knife-E_펜칼': '펜칼',
    'TS_Color_위해물품_Pistol_모형 권총(금속 재질)': '금속 권총',
    'TS_Color_위해물품_Pistol_모형 권총(플라스틱 재질)': '플라스틱 권총',
    'TS_Color_위해물품_Rifle_모형 소총(플라스틱 재질)': '플라스틱 소총',
    'TS_Color_위해물품_Scissors-A_문구용 가위(접은 형태)(037)': '가위',
}

# 클래스 레이블 매핑의 키를 NFC로 정규화
class_label_mapping_normalized = {unicodedata.normalize('NFC', k): v for k, v in class_label_mapping.items()}
```

```python
# 모델 정의 및 컴파일
def build_model(hp):
    # 이미지 입력
    image_input = Input(shape=(224, 224, 3), name='image_input')
    # 밀도값 입력
    density_input = Input(shape=(1,), name='density_input')

    # 베이스 모델 (EfficientNetB0)
    base_model = EfficientNetB0(input_tensor=image_input, include_top=False, weights='imagenet')
    base_model.trainable = True  # 모든 레이어를 학습 가능하게 설정

    # 특정 레이어까지 동결 (EfficientNetB0는 레이어 구조가 다름)
    for layer in base_model.layers[:20]:
        layer.trainable = False

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)

    # 밀도값과 합치기
    x = Concatenate()([x, density_input])

    x = Dense(
        units=hp.Int('units_dense1', min_value=256, max_value=1024, step=256),
        activation='relu'
    )(x)
    x = Dropout(hp.Float('dropout_rate1', min_value=0.3, max_value=0.7, step=0.1))(x)
    x = Dense(
        units=hp.Int('units_dense2', min_value=128, max_value=512, step=128),
        activation='relu'
    )(x)
    x = Dropout(hp.Float('dropout_rate2', min_value=0.3, max_value=0.7, step=0.1))(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=[image_input, density_input], outputs=outputs)

    optimizer = tf.keras.optimizers.Adam(
        learning_rate=hp.Float('learning_rate', min_value=1e-5, max_value=1e-3, sampling='log')
    )

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
# 튜너 설정
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=10,
    factor=3,
    directory='tuner_dir',
    project_name='efficientnetb0_tuning_with_density'  # 프로젝트 이름 변경
)
```

```python
# 콜백 정의
stop_early = EarlyStopping(monitor='val_loss', patience=2)
```

```python
# 튜닝 수행
tuner.search(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    epochs=10,
    callbacks=[stop_early]
)
```

```python
# 최적의 하이퍼파라미터로 모델 빌드
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
최적의 하이퍼파라미터:
- units_dense1: {best_hps.get('units_dense1')}
- dropout_rate1: {best_hps.get('dropout_rate1')}
- units_dense2: {best_hps.get('units_dense2')}
- dropout_rate2: {best_hps.get('dropout_rate2')}
- learning_rate: {best_hps.get('learning_rate')}
""")

# 최적의 하이퍼파라미터로 모델 생성
model = tuner.hypermodel.build(best_hps)
```

```python
# 콜백 설정
checkpoint_cb = ModelCheckpoint("/content/drive/MyDrive/Colab Notebooks/checkpoints/model_efficientnetb0_with_density.h5", save_best_only=True)
reduce_lr_cb = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
early_stopping_cb = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# 모델 훈련 (클래스 가중치 적용)
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    epochs=50,
    callbacks=[checkpoint_cb, reduce_lr_cb, early_stopping_cb],
    class_weight=class_weight_dict  # 클래스 가중치 적용
)
```

```python
# 학습 곡선 시각화
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='훈련 정확도')
plt.plot(history.history['val_accuracy'], label='검증 정확도')
plt.title('훈련 및 검증 정확도')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='훈련 손실')
plt.plot(history.history['val_loss'], label='검증 손실')
plt.title('훈련 및 검증 손실')
plt.legend()

plt.show()
```

```python
# 모델 평가
test_loss, test_acc = model.evaluate(test_generator, steps=test_steps)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_acc}")
```

```python
# 예측 수행
y_pred_probs = model.predict(test_generator, steps=test_steps)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_true_classes = []
for i in range(test_steps):
    (_, _), batch_y = test_generator[i]
    y_true_classes.extend(np.argmax(batch_y, axis=1))
y_true_classes = np.array(y_true_classes)
```

```python
# 클래스 인덱스와 클래스 레이블 매핑
inv_class_indices_mapped = {k: class_label_mapping_normalized.get(unicodedata.normalize('NFC', inv_class_indices[k]), inv_class_indices[k]) for k in inv_class_indices}
```

```python
# Classification Report 출력
target_names = [inv_class_indices_mapped[i] for i in range(num_classes)]
report = classification_report(
    y_true_classes,
    y_pred_classes,
    target_names=target_names,
    digits=4
)
print(report)
```

```python
# 혼동 행렬 계산 및 시각화
cm = confusion_matrix(y_true_classes, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
plt.figure(figsize=(12, 12))
disp.plot(cmap=plt.cm.Blues)
plt.xticks(rotation=90)
plt.show()
```

```python
# ROC 곡선 및 AUC 값 계산
from sklearn.preprocessing import label_binarize

n_classes = num_classes
y_true_bin = label_binarize(y_true_classes, classes=range(n_classes))
y_pred_probs = y_pred_probs[:len(y_true_bin)]  # 배열 길이 맞추기 (필요한 경우)

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# ROC 곡선 시각화
plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'{target_names[i]} (AUC = {roc_auc[i]:0.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('다중 클래스에 대한 ROC 곡선')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='best')
plt.show()
```

```python
# 맞춘 이미지와 틀린 이미지 시각화
# 테스트 데이터에서 이미지와 레이블 추출
x_test = []
y_test_classes = []

# 테스트 제너레이터를 순회하여 이미지와 레이블 수집
for i in range(test_steps):
    (batch_x, _), batch_y = test_generator[i]
    x_test.extend(batch_x)
    y_test_classes.extend(np.argmax(batch_y, axis=1))

# NumPy 배열로 변환
x_test = np.array(x_test)
y_test_classes = np.array(y_test_classes)

# 맞춘 샘플과 틀린 샘플의 인덱스 찾기
correct_indices = np.nonzero(y_pred_classes == y_true_classes)[0]
incorrect_indices = np.nonzero(y_pred_classes != y_true_classes)[0]

# 클래스 인덱스를 클래스명으로 변환
predicted_class_names = [inv_class_indices_mapped[i] for i in y_pred_classes]
true_class_names = [inv_class_indices_mapped[i] for i in y_true_classes]
```
